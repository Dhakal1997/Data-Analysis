Group By
the groupby() function in pandas is used to group data based on one or more columns in a dataframe

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
s_Qty = df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=s_Qty)
```

Note: most of the Qty buys Medium sizes

```python
# figure size
plt.figure(figsize=(10,5))
ax =sns.countplot( data =df, x='Courier Status',hue='Status')
plt.show()
```

```python
#histogram
df['Size'].hist()
```

```python
df['Category']= df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins =30,edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```

Note: Most of the buyers are t-shirt buyers.

```python
#checking B2B data by using pie chart
B2B_check = df['B2B'].value_counts()

#plot the pie chart
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.2f%%')
plt.show()
```

```python
#checking B2B data by using pie chart
B2B_check = df['B2B'].value_counts()

#plot the pie chart
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.2f%%')
plt.show()
```

99.24% of buyers are retailers and 0.76% are B2b buyers


```python
#prepare data for scatter plot
x_data = df['Category']
y_data = df['Size']

#plot the scatter plot
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#plt count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(x='ship-state',data=df)
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
#top 10_States
top_10_state = df['ship-state'].value_counts().head(10)
#plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data =df[df['ship-state'].isin(top_10_state.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

Note:Most of the buyers are from Maharastra

Conclusion:The data analysis reveals that the business has a significant amount of customers are from Maharastra mainly serves retailers, fulfils order through Amazon, experiences high demand for T-shirts and sees M- size as the prefered choice among buyers.
